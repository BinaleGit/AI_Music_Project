```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt

def modify_color(image, red_scale=1.0, green_scale=1.0, blue_scale=1.0):
    """
    Modify the color of an image by scaling the RGB values.

    Parameters:
    - image: NumPy array representing the image.
    - red_scale: Scaling factor for the red channel (default is 1.0).
    - green_scale: Scaling factor for the green channel (default is 1.0).
    - blue_scale: Scaling factor for the blue channel (default is 1.0).

    Returns:
    - Modified image with adjusted color.
    """
    modified_image = image.copy()  # Create a copy to avoid modifying the original image
    
    # Apply scaling to RGB channels
    modified_image[:, :, 0] = np.clip(image[:, :, 0] * red_scale, 0, 255).astype(np.uint8)  # Red channel
    modified_image[:, :, 1] = np.clip(image[:, :, 1] * green_scale, 0, 255).astype(np.uint8)  # Green channel
    modified_image[:, :, 2] = np.clip(image[:, :, 2] * blue_scale, 0, 255).astype(np.uint8)  # Blue channel
    
    return modified_image

# Define image URLs
img = 'https://raw.githubusercontent.com/narnav/data_ml/81c6ddd733880c526c155d9b3532d8c01bfc056d/dog.jpg'
img2 = 'https://raw.githubusercontent.com/BinaleGit/photos/main/hero.png'
img3 = 'https://raw.githubusercontent.com/BinaleGit/photos/main/mario_standing.png'

# Load images
pt = io.imread(img)
pt2 = io.imread(img2)
pt3 = io.imread(img3)

# Apply transformations to images
pt_flipped = pt[::-1]  # Flip vertically
pt2_flipped = pt2[:, ::-1]  # Flip horizontally

# Modify color of third image
modified_pt3 = modify_color(pt3, red_scale=1.5, green_scale=0.8, blue_scale=0.8)

# Create subplots for each image
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Display first image
axs[0, 0].imshow(pt_flipped)
axs[0, 0].set_title('Dog Image (Flipped Vertically)')

# Display second image
axs[0, 1].imshow(pt2_flipped)
axs[0, 1].set_title('Hero Image (Flipped Horizontally)')

# Display modified third image
axs[1, 0].imshow(modified_pt3)
axs[1, 0].set_title('Modified Mario Image')

# Display original third image
axs[1, 1].imshow(pt3)
axs[1, 1].set_title('Original Mario Image')

# Adjust spacing and display the plots
plt.tight_layout()
plt.show()
```
